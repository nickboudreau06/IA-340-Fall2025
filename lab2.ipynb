## Clean the Data 

Some of the columns had names that ended with a number in apostrophes, so for those columns, I renamed them to exclude the number, as it was unnecessary. I also renamed "Slot Machine Contributions to the State of Connecticut (1) (6)" to just "Contributions to the State of Connecticut". Again, there were no null values, so I had nothing to fix there. 

```python
df = df.rename(columns={'Total Contributions (1)' : 'Total Contributions', 'eBonus Contribution (3)' : 'eBonus Contribution', 'eBonus Credits Redeemed at Slot Machines (2)' : 'eBonus Credits Redeemed', 'Slot Machine Contributions to the State of Connecticut (1) (6)' : 'Contributions to the State of Connecticut'})
df.head()
```

```python
df.isnull().sum()
```

```python
df['date_only'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p').dt.date

df['date_only'] = pd.to_datetime(df['date_only'], format='%d/%m/%Y') 
df.info()
```

## Visualize the Data 


```python
sum_contributions_by_date = df.groupby('date_only')['Total Contributions'].sum() 
sum_contributions_by_date.head()
```

```python
sum_contributions_by_date.plot(title=('Total Contributions Over Time'), xlabel=('date'), ylabel=('Total Contributions'))
```

For the line plot above, I first grouped the total contributions by date, and then plotted those groups as seen above. The plot can give valuable insights into what months are performing better than others becasue the data in the dataset is entered monthly and not daily, therfore is more usuful for long term trends. However the limitation with it currently is the lack of dates covered by the data. The data can also be skewed if there is no data for one of the casinos for one of the months. 


```python
win_data = df[df['Casino']=='Foxwoods']

win_data.reset_index().plot.scatter(
    x='date_only',
    y='Win (9)',
    c='Total Contributions',
    xlabel='Date',
    ylabel='Amount Won',
    title='Wins and Contributions by Date',
    cmap='viridis'
)
```

This scatteplot combines the amount won by date with the total contributions as the color of the dot. This graph clearly shows that the total contributions and the amount won have a positive relationship. It also shows that there is an increase in the amount won from Februrary to August.
